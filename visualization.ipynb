# 파이썬 그래픽 도구 사용하기

```python
import pandas as pd    # pandas package import

# 데이터프레임의 컬럼명 지정

col = ["꽃받침길이", "꽃받침너비", "꽃잎길이", "꽃잎너비", "붓꽃종류"]

# 웹에 있는 CSV 파일 읽어와서 데이터프레임 df로 넣는다.

df = pd.read_csv("https://raw.githubusercontent.com/ykkim77/bigdata_analysis_6th/main/iris.csv", names = col)

df
```

- 그래픽 도구 활용

```python
# 선 그래프
df_new = df['꽃받침길이']
df_new.plot()
```

```python
df_new.plot(kind='bar')
```

- 상관계수 구하기
